* Bir dairenin evresini ve alann bulmak iin bir dizi ilem yapalm.

```python
pi_sayisi = 3.14
yaricap = 5
# alan
daire_alan = pi_sayisi*(yaricap**2) # yaricap**2 ifadesi yaricapi karesidir 
daire_alan
```

* Dairenin alann hesaplamak iin fonksiyon tanmlayalm.

```python
# def = definition
# daireAlanHesapla = fonksiyon ismi
# r = girdi parametresi
def daireAlanHesapla(r):
    """
        Dairenin alann hesaplar
        Girdi = Yar ap
        kt = Daire Alan
    """
    pi = 3.14
    daire_alan = pi*(r**2)
    return daire_alan
daireAlanHesapla(5)
```

* emberin evresini hesaplamak iin fonksiyon tanmlayalm.

```python
def cemberCevreHesapla(r):
    """
        emberin cevresini hesaplar
        Girdi = Yar ap
        kt = emberin evresi
    """
    pi = 3.14
    cember_cevre = 2*pi*r
    return cember_cevre
cemberCevreHesapla(4)
```

#### Lambda Fonksiyonu

### Liste Kavrama
* Liste kavrama veri analizinde veriyi ilemek iin kullanlan etkili bir yntemdir.

```python
# veriyi olutur
import pandas as pd
sozluk = {"isim": ["ali","veli","kenan","murat","ayse","hilal"],
              "yas" : [15,16,17,33,45,66],
              "sehir": ["zmir","Ankara","Konya","Ankara","Ankara","Antalya"]} 
veri = pd.DataFrame(sozluk)
veri
```

```python
veri["YAS_GRUBU"] = ["kucuk" if ortalama_yas > i else "buyuk" for i in veri.yas]
veri
```

### Histogram Grafii
* Histogram, gruplandrlm bir veri dalmnn stun grafiiyle gsterimidir. Dier bir ifadeyle, tekrarl saylardan oluan verilerin, uygulanan ilemlerden sonra nce tabloya, tablodan yararlanarak grafie aktarlmas, yani veri gruplarnn grafiinin dikdrtgen stunlar halinde gsterilmesidir. 

```python
# Ktphaneleri ie aktar ve veriyi ykle
import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("iris.csv")
veri.head()
```

```python
# imdi sepal length ve petal length stunlarnn histogram izdirelim
plt.figure() # figr olutur - d ereve
plt.hist(veri["SepalLengthCm"], color = "green", alpha = 0.5, bins = 20, label = "Alt Yaprak Uzunluk") # alpha = saydamlk, bins = bar says
plt.hist(veri["PetalLengthCm"], color = "orange", alpha = 0.5, bins = 20, label = "st Yaprak Uzunluk") # alpha = saydamlk, bins = bar says
plt.title('Alt Yaprak Uzunluu ve st Yaprak Uzunluu - Histogram')
plt.xlabel("Cm")
plt.ylabel("Frekans")
plt.legend() # label'larn grnebilmesi iin
plt.show()
```

### Alt Plan Grafii
* Alt plan grafiklerinde birden fazla grafik tek bir figr/ereve zerinde bulunur.

```python
# Ktphaneleri ie aktar ve veriyi ykle
import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("iris.csv")
veri.head()
```

```python
fig, ax = plt.subplots(2, 1) # figr zerine 2*1 lik bir tablo olutur. bu tablo zerine ax (eksen) ekl
fig.suptitle("Alt Yaprak Uzunluk ve st Yaprak Uzunluk", fontsize=14)
ax[0].plot(veri.Id,veri.SepalLengthCm,color="red",label= "SepalLengthCm")
ax[0].set_title('Alt Yaprak Uzunluk') # balk
ax[0].set_xlabel("id") # x ekseni label
ax[0].set_ylabel("Cm") # y ekseni label

ax[1].plot(veri.Id,veri.PetalLengthCm,color="green",label= "PetalLengthCm")
ax[1].set_title('st Yaprak Uzunluk')
ax[1].set_xlabel("id")
ax[1].set_ylabel("Cm")

fig.tight_layout() # figr aras yerleimi ayarla
fig.subplots_adjust(top=0.85)
plt.show()
```
